```python
def flist(file):
    with open(file) as f:
        lines = f.read().splitlines()
    return lines
def split(f):
    header = f[:8]
    polys = f[8:-2]
    foot = f[-2:]
    return header, polys, foot
def poly_in(p,start,n):
    off = start*5
    lst = []
    while(off<n*5):
        lay = p[off+1]
        num = p[off+3][4]
        cord = p[off+3][7:]
        off+=5
#         print(lay)
#         print(num)
#         print(cord)
        temp = []
        temp.append(lay)
        temp.append(num)
        temp.append(cord)
        lst.append(temp)
    return lst
def poly_out(poly_lst):
    off = 0
    n = len(poly_lst)
    out = []
    while(off<n):
        out.append('boundary')
        out.append(poly_lst[off][0])
        out.append('datatype 0')
#         print(poly_lst[off])
        out.append('xy  '+poly_lst[off][1]+'  '+poly_lst[off][2])
        out.append('endel')
#         out.append()
        off+=1
    print(out)
    return out  
def lst_file(header,poly_op,foot):
    out_lst = header+poly_op+foot
    print(out_lst)
    with open(r'milestone3.txt', 'w') as fp:
        for item in out_lst:
            # write each item on a new line
            fp.write("%s\n" % item)
        print('Done')
```
